```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# array form -1 to 1
x = np.linspace(-1,1,201)
x
```

```python
# arange requires start, end, step
dt = 0.01
tMin = -1
tMax = 1
t = np.arange(tMin, tMax + dt, dt)
t
```

```python
print(x.shape, t.shape)
```

```python
x == t
```

```python
for i in range(201):
    print(i, x[i]==t[i])
```

```python
y = np.zeros(x.shape)
z = np.zeros(t.shape)

# fill y using a for loop
for i in range(201):
    y[i] = x[i]-x[i]**3

# fill z using vectorized math
# this will list results one iteration at a time
z = t - t**3
```

```python
plt.plot(x,y, 'r-', lw=2, label='for loop');
plt.plot(t,z, 'b--', lw=2, label='array math');
plt.legend();
```
