Every time you call `rand()`, Python generates a random number between 0 and 1.  We can use Python's logic functions to turn this into a binary function.  Run the cell below several times.

Describe what the output, and how it might be used to simulate the flip of a coin.

-----

The output of rand() < 0.5 gives us a boolean result of whether or not the randomly generated number between 0 and 1 would be less than 0.5. This is a good representation of a coin flip because the computer randomly generates a number (the flip of the coin) and if it generates a number less than 0.5 (Lets say this is heads) then the boolean result is "False" and if the generated number is more than 0.5 (tails) then the boolean result is printed as "True"

-----

The goal of this problem is to "generate 6000 six-bit random binary fractions" and plot the result as a histogram.  Thus, we need to flip a coin six times, then use the results of each flip to create a binary fraction.  Let $s_n$ be the outcome of the $n$-th flip.  Then, the number generated by a set of six flips is

$$x = \dfrac{s_0}{2} + \dfrac{s_1}{2^2} + \dfrac{s_2}{2^3} + \dfrac{s_3}{2^4} + \dfrac{s_4}{2^5} + \dfrac{s_5}{2^6}$$

(This uses Python indexing for the coin flips.)

Run the following cell several times.  Verify that it generates a six-bit random binary fraction, then explain how it works.  That is, parse each line an explain how the Python code is implementing the matematical formula above.

-----

```python
s = rand(6) < 0.5 #generates six random numbers and assigns a boolean value of 'True' to it depending on if it is less than (<) 0.5, or else it is assigned 'False'
x = np.sum(s / 2**np.arange(1,7)) #generates a six bit fraction using the mathematical formula above

print("s= {}".format(s)) #prints the boolean values of the six randomly generated numbers
print("x= {}".format(x)) #prints the six bit fraction
```

Yes, this code generates a six-bit random binary fraction and assigns it to the variable 'x' in the second line of code. This is accomplished by generating the simulated flip of a coin six times in the first line of code, and assigning it to the variable 's' in boolean format that tells us whether that flip was heads (False) or tails (True) by determining if the random number generated, between 0 and 1, will be less than (<) 0.5. Then in the second line of code, in order to create the binary fraction, we divide the six coin flips by 2 to the power of 'n' flips, for their respective flip count, and sum the six fractions. Then the last two lines of code display the values assigned to variables 's' and 'x', which are the matrix vector containing the 6 coin flips and the six bit randomly generated fraction, respectively

------

Now, write a `for` loop that will create an array and fill it with `N=6,000` six-bit random binary fractions.  (You can copy and paste the first two lines of the code cell above into the body of the `for` loop, or develop your own algorithm.)

```python
N = 6000
X = np.zeros(N) #creates an array consisting of 6000 zeros
for n in range(N):
    s = rand(6) < 0.5
    x = np.sum(s / 2**np.arange(1,7)) #generates a six bit fraction using the mathematical formula above
    X[n] = x #assigns each six bit fraction generated in each iteration to its corresponding index in array X
print("The 6000 random binary fractions are {}".format(X)) #prints the array consisting of the six bit fraction created in each iteration of the 'for' loop
```

Use `plt.hist` to create a histogram of your data set.

```python
plt.hist(X);
```
