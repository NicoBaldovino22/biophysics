## (a) __Histogram of $1/x$__

```python
#Standard import of numpy and matplotlibrary
import numpy as np;
import matplotlib.pyplot as plt;

#Random Number Generator
from numpy.random import default_rng;
rand = default_rng().random; #assigns a random number to variable 'rand' in the range of 'rng' as default range from 0 to 1

# Import binomial coefficient function
from scipy.special import binom
```

```python
# Size of the problem and base probability, because we want a uniform distibution, xi = 0.5
M = 10
xi = 0.5

# Array of all possible outcomes
k = np.arange(M+1)

# Array of probabilities, created with array math for binomial distributions
Pk = binom(M,k) * xi**k * (1-xi)**(M-k)

# Bin locations for simulation.  First entry needs to be zero.
bins = np.zeros(M+2)
bins[1:] = Pk.cumsum()

# Generating a random data set of 10000 uniformly distributed random numbers
N = 10000
x = rand(N)
y = 1/x
print(x)
print(y)
plt.figure(1)
plt.hist(x,bins=bins)
plt.figure(2)
plt.hist(y,bins=k)
```

## (b) Log-log plot of frequencies

```python
counts, bins = np.histogram(x, bins=bins)
plt.figure(3)
plt.loglog(counts,'r-',label='observed frequencies')
plt.title('observed frequency of drawn value for 10,000 \n randomly generated and uniform numbers')
plt.legend()
```
