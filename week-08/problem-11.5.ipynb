## (a) Plot the phase portrait.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# Creating the vector field for the phase portrait
t = np.linspace(-3,3,21)
X, Y = np.meshgrid(t,t)
U = -Y - 0.1 * X
V = X - 0.1 * Y

# Generating a graphical representation of this phase portrait
plt.figure(figsize=(12,12))
plt.quiver(X,Y,U,V, pivot='mid') # pivot makes sure that the vector field aligns in the middle of the plot

# Implementing the equation of motion using the illistrative parameter values g/L = 1 and xi/m = 1/3
m = 3
g = 1
xi= 1
L = 1
o = 0

theta = np.linspace(-np.pi,np.pi,101)
y = np.linspace(0, 40, 101)
X,Y= np.meshgrid(theta,y)

# Equation of Motion
o_time = -((m*g*np.sin(theta)) + (xi*L*o))/ (m*L)
plt.plot(theta,o_time, color='orange')
plt.axis('equal')

plt.plot([-np.pi,np.pi],[0,0],color='orange')

plt.plot(0,0,'og', ms=20)
plt.plot(-np.pi, 0, 'or', ms=20)

# Creating the stream line funtion that would be generated by the swinging pendulum
def F(y,t):
    dy = [0,0]
    dy[0] = -y[1] - 0.1 * y[0]
    dy[1] = y[0] - 0.1 * y[1] 
    return dy

# Total elapsed time array
T = np.linspace(0,40,1001)

# Streamline 1 of the first pendulum trajectory (Un-Damped)
y0 = [-np.pi,0]
y = odeint(F, y0, T)

plt.plot(y[:,0],y[:,1],color='black')
plt.axis('equal')

# Streamline 2 of the second pendulum trjectory (Un-Damped)
y0 = [np.pi,0]
y = odeint(F, y0, T)

plt.plot(y[:,0],y[:,1],color='black')
plt.axis('equal')
```
